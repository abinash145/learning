```python
from tkinter import *
calcu = Tk()
calcu.title("simple calculator")
data = Entry(calcu,text="")
data.grid(row=0,column=0,columnspan=3,pady=10,padx=10)


def click(a):
    ram=data.get()
    data.delete(0,END)
    data.insert(0,f"{ram}{a}")
    return
def add(a):
    num=data.get()
    data.delete(0,END)
    global f1
    f1=int(num)
    data.delete(0,END)
    global n
    n=a
    return      
    
def ans(a):
    f2=data.get()
    data.delete(0,END)
    if a==1:
        answer=int(f1)+int(f2)
        data.insert(0,answer)
    elif a==2:
        answer=int(f1)-int(f2)
        data.insert(0,answer)
    elif a==3:
        answer=int(f1)*int(f2)
        data.insert(0,answer)
    else :
        answer=int(f1)/int(f2)
        data.insert(0,answer)   
   
    return
def cleara():
    data.delete(0,END)

button1 = Button(calcu,text="1",padx=40,pady=40,command=lambda:click(1))
button2 = Button(calcu,text="2",padx=40,pady=40,command=lambda:click(2))
button3 = Button(calcu,text="3",padx=40,pady=40,command=lambda:click(3))
button4 = Button(calcu,text="4",padx=40,pady=40,command=lambda:click(4))
button5 = Button(calcu,text="5",padx=40,pady=40,command=lambda:click(5))
button6 = Button(calcu,text="6",padx=40,pady=40,command=lambda:click(6))
button7 = Button(calcu,text="7",padx=40,pady=40,command=lambda:click(7))
button8 = Button(calcu,text="8",padx=40,pady=40,command=lambda:click(8))
button9 = Button(calcu,text="9",padx=40,pady=40,command=lambda:click(9))
button0 = Button(calcu,text="0",padx=40,pady=40,command=lambda:click(0))
clearb = Button(calcu,text="clear",padx=40,pady=40,command=cleara)
equal = Button(calcu,text="=",padx=40,pady=40,command=lambda:ans(n))
plus = Button(calcu,text="+",padx=40,pady=40,command=lambda:add(1))
minus = Button(calcu,text="-",padx=40,pady=40,command=lambda:add(2))
multiple = Button(calcu,text="ร",padx=40,pady=40,command=lambda:add(3))
divide = Button(calcu,text="รท",padx=40,pady=40,command=lambda:add(4))


button1.grid(row=1,column=1)
button2.grid(row=1,column=2)
button3.grid(row=1,column=3)
button4.grid(row=2,column=1)
button5.grid(row=2,column=2)
button6.grid(row=2,column=3)
button7.grid(row=3,column=1)
button8.grid(row=3,column=2)
button9.grid(row=3,column=3)
button0.grid(row=4,column=1)
plus.grid(row=4,column=2)
minus.grid(row=4,column=3)
multiple.grid(row=5,column=1)
divide.grid(row=5,column=2)
equal.grid(row=5,column=3)
clearb.grid(row=6,column=1,columnspan=3)



calcu.mainloop()
```
